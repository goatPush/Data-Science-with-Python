```python
# Month to month trends for items sold in months Jan-Jun
import numpy as np
from matplotlib import pyplot as plt

# Create x-axis values
x_months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

# Create mock items sold per month numbers
# Values ranging from 1,000 to 2,000 in 200 unit increments)
y_items_per_month = np.array(range(1000, 2000 + 1, 200))
```

```python
# Print off axes values
print('x values:')
for i in x_months:
    print(f'  {i}')
    
print('\ny values:')
for i in y_items_per_month:
    print(f'  {i}')
```

```python
# Generate a graph:

# Chart title
plt.title('Items Sold has been Increasing Linearly:',fontdict={'fontsize': 22, 
                                                               'fontweight': 'bold'})

# Axes labels
plt.xlabel('Month')
plt.ylabel('Items Sold')

# Plot line: Blue dotted, with star markers
plot_kwargs = {'markersize':10, 
               'markerfacecolor': 'red',
               'markeredgecolor': 'darkred'
              }
plt.plot(x_months, y_items_per_month, '*:b', **plot_kwargs)
# Draw plot
plt.show()
```
