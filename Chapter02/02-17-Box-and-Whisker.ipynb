```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, zscore
```

```python
# Create normally distributed values
y_values = np.random.normal(loc=0, scale=0.1, size=100)
```

```python
# Plot values on a histogram
plt.boxplot(y_values)
plt.show()
```

```python
# S-W values
shap_w, shap_p = shapiro(y_values)

# Get z scores of y values
y_z_scores = zscore(y_values)
```

```python
# Count the number of outliers
total_outliers = 0
for i in range(len(y_z_scores)):
    if abs(y_z_scores[i]) >= 3:
        total_outliers += 1
```

```python
# Create and set title
if shap_p > 0.05:
    title = f'Normally distributed with {total_outliers} outlier(s)'
else:
    title = f'Not normally distributed with {total_outliers} outliers(s)'
    
plt.title(title)

# Plot values on a histogram
plt.boxplot(y_values)
plt.show()
```
