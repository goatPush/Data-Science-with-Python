```python
import numpy as np
import matplotlib.pyplot as plt

# Generate array with values 20 evenly spaced values from 0 to 10
x_ndarray = np.linspace(0, 10, 20)
```

```python
# Create array with cubes of x_ndarray
y_ndarray = x_ndarray ** 3
```

```python
# Plot as a curve using matplotlib
plt.plot(x_ndarray, y_ndarray)
plt.show()
```

```python
# Add axis labels and chart title
plt.xlabel('Linearly Spaced numbers')
plt.ylabel('f(x) = x')
plt.title('x by x Cubed')

# Change line color
plt.plot(x_ndarray, y_ndarray, 'k')
plt.show()
```

```python
# Add axis labels and chart title
plt.xlabel('Linearly Spaced numbers')
plt.ylabel('f(x) = x')
plt.title('x by x Cubed')

# Plot with black diamond data markers
plt.plot(x_ndarray, y_ndarray, 'Dk')
plt.plot()
```

```python
# Add axis labels and chart title
plt.xlabel('Linearly Spaced numbers')
plt.ylabel('f(x) = x')
plt.title('x by x Cubed')

# Plot with line and markers
plt.plot(x_ndarray, y_ndarray, 'D-k')
plt.show()
```

### ***Note: Font args have been moved to a dictionary called fontdict***

```python
# Add axis labels and chart title
plt.xlabel('Linearly Spaced numbers')
plt.ylabel('f(x) = x')

# Increase title font
plt.title('x by x Cubed', fontdict={'fontsize':22, 
                                    'fontweight':'bold', 
                                    'color':'blue'})

# Plot with line and markers
plt.plot(x_ndarray, y_ndarray, 'D-k')
plt.show()
```
