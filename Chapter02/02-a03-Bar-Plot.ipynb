```python
"""
Make a bar plot that compares the number of NBA championships
among five franchises with the most titles.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Top 5 winningest NBA teams
x_top_franchises = np.array(['LA Lakers', 
                             'Boston Celtics', 
                             'Golden State Warriors', 
                             'Chicago Bulls', 
                             'SA Spurs'])

# Number of wins each
y_championships = np.array([17, 17, 7, 6, 5])

# Create dataframe out of lists
df = pd.DataFrame({'Team': x_top_franchises, 'Titles': y_championships})

# Sort by Titles descending
df.sort_values('Titles', ascending=False, inplace=True)
df
```

```python
# Get teams with the most wins
winningest_teams = np.array(df.loc[(df['Titles'] == df['Titles'].max())].get('Team'))

# Construct chart title
title = ''
have_has = 'have'
for i in range(0, len(winningest_teams)):
    title += f'The {winningest_teams[i]}'
    if len(winningest_teams) > 1:
        if i <= len(winningest_teams) - 2 and len(winningest_teams) > 2:
            title += ','
        if i == len(winningest_teams) - 2:
            title += ' and'
    else: 
        have_has = 'has'
    # end outer if/else
        
    title += ' '
#end for
        
title += f'{have_has} won the most NBA championships'
title
```

```python
# Set tile and axes labels
plt.title(title, fontdict={'fontsize':12, 'fontweight': 'bold'})

plt.xlabel('NBA Franchises')
plt.ylabel('Number of Championships')

#Rotate ticks
plt.xticks(rotation=45)

# Draw plot
bar_kwargs = {'color': 'darkred'}
plt.bar(df['Team'], df['Titles'], **bar_kwargs)

# Save plot
plt.savefig('Titles-won-by-NBA-franchises-top-5', bbox_inches='tight')

plt.show()
```
