```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Create random, normally distributed values with 0 mean
y_values = np.random.normal(loc=0, scale=0.1, size=100)
```

```python
# Create histogram
plt.hist(y_values, bins=20, rwidth=0.95)

# Label axes
plt.xlabel('y Value')
plt.ylabel('Frequency')

plt.show()
```

```python
# Use Shapiro-Wilk test to verify data is normally distributed
shap_w, shap_p = shapiro(y_values)

# Create dynamic chart title
normal_YN = ''
if shap_p > 0.05:
    normal_YN = 'Fails to reject S-W null hypothesis\nData is normally distributed'
else:
    normal_YN = 'S-W null hypothesis is rejected\nData is not normally distributed'

# Set title
plt.title(normal_YN, fontdict={'fontsize': 12, 'fontweight': 'bold'})

# Create histogram
plt.hist(y_values, bins=20, rwidth=0.95)

# Label axes
plt.xlabel('y Value')
plt.ylabel('Frequency')

plt.show()
```
