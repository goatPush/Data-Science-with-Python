```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Create list of numbers representing height
y_heights = np.array([5, 5.5, 5, 5.5, 6, 6.5, 6, 6.5, 7, 5.5, 5.25, 6, 5.25])

# Create list of numbers representing weight
x_weights = np.array([100, 150, 110, 140, 140, 170, 168, 165, 180, 125, 115, 155, 135])
```

```python
# Create scatterplot
plt.scatter(x_weights, y_heights)

# Label axes
plt.xlabel('Weight')
plt.ylabel('Height')

plt.show()
```

```python
# Find the strength of correltation
correlation_coeff, p_value = pearsonr(x_weights, y_heights)

# Build title string
title = 'There is a'
if np.abs(correlation_coeff) >= 1: #Should never be >1, but...authors' logic needs to be fixed
    title += ' perfect,'
elif np.abs(correlation_coeff) >= 0.8:
    title += ' very strong,'
elif np.abs(correlation_coeff) >= 0.6:
    title += ' strong,'
elif np.abs(correlation_coeff) >= 0.4:
    title += ' moderate,'
elif np.abs(correlation_coeff) >= 0.2:
    title += ' weak,'
elif np.abs(correlation_coeff) > 0:
    title += ' very weak,'
else:
    title += ' no'

if correlation_coeff > 0:
    title += ' postitive'
elif correlation_coeff < 0:
    title += ' negative'
    
title += f' linear relationship (r = {correlation_coeff:0.2f})'
title
```

```python
# Set chart title
plt.title(title, fontdict={'fontweight': 'bold'})

# Create scatterplot
plt.scatter(x_weights, y_heights)

# Label axes
plt.xlabel('Weight')
plt.ylabel('Height')

plt.show()
```
