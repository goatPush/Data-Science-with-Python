```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
 # Import data sets
items_dataset = 'Items_Sold_by_Week.csv'
items_df = pd.read_csv(items_dataset, header=0)

wh_dataset = 'Weight_by_Height.csv'
wh_df = pd.read_csv(wh_dataset, header=0)
```

```python
# Create normally distributed set of numbers (see ex 17)
norm_dist_nums = np.random.normal(loc=0, scale=0.1, size=100)
```

```python
# Set tight layout
plt.tight_layout()

# Create figure and axes sets in  2x3 grid
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,10), layout='constrained')

# Name the axes
[[line_splt, bar_splt], 
      [hbar_splt, histo_splt], 
      [scatter_splt, baw_splt]] = axes
```

```python
# Set titles and label axes
line_splt.set_title('Line')
bar_splt.set_title('Bar')
hbar_splt.set_title('Horizontal Bar')
histo_splt.set_title('Histogram')
scatter_splt.set_title('Scatter')
baw_splt.set_title('Box-and-Whisker')

# Label axes
week_axis_title = 'Week'
items_axis_title = 'Items sold'
norm_dist_axis_title = 'Normally distributed values'

line_splt.set_xlabel(week_axis_title)
bar_splt.set_xlabel(week_axis_title)
hbar_splt.set_xlabel(items_axis_title)

line_splt.set_ylabel(items_axis_title)
bar_splt.set_ylabel(items_axis_title)
hbar_splt.set_ylabel(week_axis_title)

histo_splt.set_xlabel(norm_dist_axis_title)
baw_splt.set_xlabel(norm_dist_axis_title)

histo_splt.set_ylabel('Count by bin')

scatter_splt.set_xlabel('Weight (lb.)')
scatter_splt.set_ylabel('Height (ft.)')
```

```python
# Plot Items sold by week on the Line, Bar, and Horizontal Bar plots
line_splt.plot(items_df['Week'], items_df['Items_Sold'])
bar_splt.bar(items_df['Week'], items_df['Items_Sold'])
hbar_splt.barh(items_df['Week'], items_df['Items_Sold'])
```

```python
# Plot normally distributed numbers in histogram and box-and-whisker
histo_splt.hist(norm_dist_nums, rwidth=0.9)
baw_splt.boxplot(norm_dist_nums, vert=False)
```

```python
# Plot the Weight and Height data in the scatter plot
scatter_splt.scatter(wh_df['Weight'], wh_df['Height'])
```
