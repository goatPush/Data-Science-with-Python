```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create x and y axes values
x_values = np.linspace(0, 10, 20)

y_cubes = x_values**3

# Let's make the second plot the derivative for fun
y_prime = 3*(x_values**2)
```

```python
# Create figure and two sets of axes
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.show()
```

```python
# Create figure and two sets of axes
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot f(x) = x
axes[0].plot(x_values, y_cubes)

# Set first axes title and labels
axes[0].set_title('x')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('f(x) = x')

# Plot f'(x) and set title and labels
axes[1].plot(x_values, y_prime)

axes[1].set_title('3x')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('f\'(x) = 3x')

# Set layout
plt.tight_layout()

# Draw
plt.show()
```
