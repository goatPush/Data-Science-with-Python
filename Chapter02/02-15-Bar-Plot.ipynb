```python
import numpy as np
import matplotlib.pyplot as plt

# Create list of categories
x_clothing_cat = np.array(['Shirt',
                           'Pants',
                           'Shorts',
                           'Shoes'])

# Create mock sales revenue numbers per category
y_revenue = np.array([1000, 1200, 800, 1800])

print(f'Clothing category revenue:')
for i in range(0, len(x_clothing_cat)):
    print(f'  {x_clothing_cat[i] + ":":<7} {y_revenue[i]: >4}')
```

```python
# Plot and show bar graph
plt.bar(x_clothing_cat, y_revenue)
plt.show()
```

```python
# Set title and axis labels
title_kwargs = {'fontsize': 22,
                'fontweight': 'bold'}
plt.title('Sales Revenue by Item Type', **title_kwargs)

plt.xlabel('Item Type')
plt.ylabel('Sales revenue ($)')

# Plot and show bar graph
plt.bar(x_clothing_cat, y_revenue)
plt.show()
```

```python
# Find category with most revenue and dynamically set the chart title
index_max_revenue = np.where(y_revenue == y_revenue.max())[0][0]

highest_revenue_cat = x_clothing_cat[index_max_revenue]

# Set title and axis labels
title_kwargs = {'fontsize': 14,
                'fontweight': 'bold'}
plt.title(f'{highest_revenue_cat} Produced the Most Sales Revenue', **title_kwargs)

plt.xlabel('Item Type')
plt.ylabel('Sales revenue ($)')

# Plot and show bar graph
plt.bar(x_clothing_cat, y_revenue)
plt.show()
```

```python
# Find category with most revenue and dynamically set the chart title
index_max_revenue = np.where(y_revenue == y_revenue.max())[0][0]

highest_revenue_cat = x_clothing_cat[index_max_revenue]

# Set title and axis labels (switch axes for horizontal bar chart)
title_kwargs = {'fontsize': 14,
                'fontweight': 'bold'}
plt.title(f'{highest_revenue_cat} Produced the Most Sales Revenue', **title_kwargs)

plt.ylabel('Item Type')
plt.xlabel('Sales revenue ($)')

# Plot and show *horizontal* bar graph
plt.barh(x_clothing_cat, y_revenue)
plt.show()
```
