```python
import numpy as np
import matplotlib.pyplot as plt

# Generate array with values 20 evenly spaced values from 0 to 10
x_ndarray = np.linspace(0, 10, 20)

# Create array with cubes of x_ndarray
y_ndarray = x_ndarray ** 3

# Create a second y array
y2_ndarray = x_ndarray **2
```

```python
# Plot y_ndarray from Exercise 12:
# Add axis labels and chart title
plt.xlabel('Linearly Spaced numbers')
plt.ylabel('f(x) = xÂ³')

# Increase title font
plt.title('x by x Cubed', fontdict={'fontsize':22, 
                                    'fontweight':'bold', 
                                    'color':'blue'})

# Plot with line and markers
plt.plot(x_ndarray, y_ndarray, 'D-k')

# Plot y2_ndarray on same graph
plt.plot(x_ndarray, y2_ndarray)
plt.show()
```
