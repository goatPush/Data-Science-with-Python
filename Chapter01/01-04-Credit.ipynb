```python
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sbn

# Get Data and make dataframe
dataset = 'Data/german_credit_data.csv'
df = pd.read_csv(dataset, header=0)
```

```python
sbn.boxplot(df['Age'], orient='h')
```

```python
# Calculate quantiles and interquantile
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Calculate lower and upper fences
Lower_Fence = Q1 - (1.5 * IQR)
Upper_Fence = Q3 + (1.5 * IQR)
print(f'Lower fence: {Lower_Fence}\n' \
     f'Upper fence: {Upper_Fence}')
```

```python
# Print outliers
df[((df['Age'] < Lower_Fence) | (df['Age'] > Upper_Fence))]
```

```python
# drop outliers from dataframe
df = df[~((df['Age'] < Lower_Fence) | (df['Age'] > Upper_Fence))]
df.head(10)
```
